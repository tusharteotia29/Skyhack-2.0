```python
import pandas as pd
import sweetviz as sv
```

```python
df_sentiment = pd.read_csv('/kaggle/input/united-hackathon-dataset/sentiment_statisticscc1e57a.csv')
df_sentiment
```

```python
df_customer=pd.read_csv('/kaggle/input/united-hackathon-dataset/customers2afd6ea.csv')
df_customer
```

```python
df_calls = pd.read_csv('/kaggle/input/united-hackathon-dataset/callsf0d4f5a.csv')
df_calls
```

```python
df_reasons = pd.read_csv('/kaggle/input/united-hackathon-dataset/reason18315ff.csv')
df_reasons
```

```python
result1 = pd.merge(df_sentiment, df_calls, on='call_id', how='left')
result1
```

```python
result2 = pd.merge(result1, df_customer, on='customer_id', how='left')
result2
```

```python
result2.describe()
```

```python
df=result2
```

```python
# Convert to datetime
df['call_start_datetime'] = pd.to_datetime(df['call_start_datetime'])

# Extract components
df['start_current_time'] = df['call_start_datetime'].dt.time
df['start_month'] = df['call_start_datetime'].dt.month
df['start_date'] = df['call_start_datetime'].dt.day
df['start_day'] = df['call_start_datetime'].dt.day_name()
df['start_year'] = df['call_start_datetime'].dt.year

df
```

```python
# Convert to datetime
df['agent_assigned_datetime'] = pd.to_datetime(df['agent_assigned_datetime'])

# Extract components
df['agent_start_current_time'] = df['agent_assigned_datetime'].dt.time
df['agent_start_month'] = df['agent_assigned_datetime'].dt.month
df['agent_start_date'] = df['agent_assigned_datetime'].dt.day
df['agent_start_day'] = df['agent_assigned_datetime'].dt.day_name()
df['agent_start_year'] = df['agent_assigned_datetime'].dt.year

df
```

```python
df['call_end_datetime'] = pd.to_datetime(df['call_end_datetime'])

# Extract components
df['end_current_time'] = df['call_end_datetime'].dt.time
df['end_month'] = df['call_end_datetime'].dt.month
df['end_date'] = df['call_end_datetime'].dt.day
df['end_day'] = df['call_end_datetime'].dt.day_name()
df['end_year'] = df['call_end_datetime'].dt.year

df
```

```python
df["AHT"]= df['call_end_datetime'] - df['agent_assigned_datetime']
df['AST']= df['agent_assigned_datetime'] - df['call_start_datetime']
df
```

```python
df["AHT_minutes"] = df["AHT"].dt.total_seconds() / 60
df["AST_minutes"] = df["AST"].dt.total_seconds() / 60
```

```python
df
```

```python
report = sv.analyze(df)
```

```python
report.show_html('sweetviz_report.html')
```
